```python
# what was the best month for sales? How much we earned that month

import pandas as pd

# Assuming `a` is your DataFrame
a['Order Date'] = pd.to_datetime(a['Order Date'], errors='coerce')  # Convert 'Order Date' column to datetime format, handling errors
a['Month'] = a['Order Date'].dt.month  # Extract the month and assign it to a new column 'Month'

# Now the 'Month' column contains the month extracted from the 'Order Date' column

```

```python
a=a.dropna(how='all')
a.dropna(axis=0,inplace=True)
a.head()
```

```python

# convert Quantity and price to Int
a['Quantity Ordered'] = pd.to_numeric(a['Quantity Ordered'], errors='coerce')
a['Price Each'] = pd.to_numeric(a['Price Each'], errors='coerce')
#a[['Quantity Ordered','Price Each']]=a[['Quantity Ordered','Price Each']].astype(int)
```

```python
# Add a Sales column 
a['Sales']=a['Quantity Ordered']*a['Price Each']
```

```python
# what time should we display advertisements to maximize likelihood of the customers buying product

a["Order Date"]=pd.to_datetime(a['Order Date'])
a["hour"]=a['Order Date'].dt.hour
a["minute"]=a['Order Date'].dt.minute
a.head()
```

```python
### At what time highest numbers of orders are placed

hours = a["hour"].unique()

# Plot a bar graph for each hour
for hour in hours:
    data_hour = a[a['hour'] == hour]
    hour_counts = a['hour'].value_counts().sort_index()
hour_counts.plot(kind='bar')
```

```python
# what products are sold often together?

a["Grouped"]=a.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

#a=a[['Order ID','Grouped']].drop_duplicates()
a.head(100)
```

```python
from itertools import combinations
from collections import Counter
c=Counter()
for row in a["Grouped"]:
    row_list=row.split(',')
    c.update(Counter(combinations(row_list,2)))
for key,value in c.most_common(10):
    print(key,value)
```

```python
# What product sold the most ? 
a.head()
pg=a.groupby('Product')
QO=pg.sum()['Quantity Ordered']
products=[p for p, a in pg]
QO.plot(kind='bar')
```

```python
prices = a.groupby('Product').mean()['Price Each']

fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,QO,color='g')
ax2.plot(products,prices,'b-')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('price($)',color='b')
ax1.set_xticklabels(products,rotation='vertical',size=8)
plt.show()
```
